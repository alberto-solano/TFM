```python
import numpy as np
import gym
import time 
from gym import spaces
import matplotlib.pyplot as plt
from env_variable_info import RegEnv
from agents import  DQN
from IPython.display import display, clear_output
import torch
import torch.nn as nn
```

```python
lr = 0.1
env = RegEnv(lr=lr)
```

```python
decay = True
alpha = 0.3
gamma = 0.7
epsilon = 1
episodes = 50000
disc_factor = 0.15
discount = 1-1/(episodes*disc_factor)
def adaptive(self, episode):
    self.epsilon = max(0.01, min(1.0, self.epsilon*discount))
```

```python
model = DQN(env, alpha, gamma, epsilon, adaptive = adaptive)
model.Q_net.load_state_dict(torch.load('weights_100000_0.001_0.99_128_True_0.35.pt'))
```

```python
state = env.reset()
done = False
while not done:
    action = model.predict(state)
    state, reward, done, _ = env.step(action)
    fig = env.render()
    fig.suptitle(action)
    display(fig)
    time.sleep(0.2)
    clear_output(wait=True)
```
